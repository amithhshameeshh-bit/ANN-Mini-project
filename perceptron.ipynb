```python
import numpy as np

# AND Gate Dataset
X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([0,0,0,1])

# Initialize
w = np.zeros(2)
b = 0
lr = 0.1

# Training
for epoch in range(10):
    for i in range(len(X)):
        z = np.dot(X[i], w) + b
        y_pred = 1 if z > 0 else 0
        error = y[i] - y_pred

        w += lr * error * X[i]
        b += lr * error

# Testing
print("Final Predictions:")
correct = 0
for i in range(len(X)):
    z = np.dot(X[i], w) + b
    y_pred = 1 if z > 0 else 0
    print(f"Input: {X[i]} → Predicted: {y_pred}, Actual: {y[i]}")
    if y_pred == y[i]:
        correct += 1

accuracy = correct / len(X) * 100
print("Accuracy:", accuracy, "%")
```
